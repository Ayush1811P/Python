```python
import matplotlib.pyplot as plt
regions=['NORTH','SOUTH','EAST','WEST']
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions, autopct='%1.1f%%',colors=['gold','skyblue','green','blue'])
plt.title("REVENUE CONTRIBUTION BY REGION ")
plt.savefig('First-file1.png',dpi=300, bbox_inches='tight')
plt.show()
```
