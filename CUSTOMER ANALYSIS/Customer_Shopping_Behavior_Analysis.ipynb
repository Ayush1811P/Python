```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# # Imputing missing values in Review Rating column with the median rating of the product category

# df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))


# ...existing code...
# Robust imputation for review rating (handles different column name variants and non-numeric values)
def _normalize(name):
    return name.lower().strip().replace(' ', '_')

cols_norm = { _normalize(c): c for c in df.columns }
review_col = cols_norm.get('review_rating') or cols_norm.get('reviewrating')
category_col = cols_norm.get('category')

if review_col is None:
    raise KeyError("Review rating column not found (expected names like 'Review Rating' or 'review_rating').")

# ensure numeric
df[review_col] = pd.to_numeric(df[review_col], errors='coerce')

# impute by category median if category exists, otherwise use overall median
if category_col:
    df[review_col] = df.groupby(category_col)[review_col].transform(lambda x: x.fillna(x.median()))
else:
    df[review_col] = df[review_col].fillna(df[review_col].median())
# ...existing code...
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
